### 2. Subset the data from dates ( DEc 2006 to Nov 2009)

```python
df['Datetime'] = pd.to_datetime(df['Date'])
a = '2006-12-01'
b = '2009-11-30'
df1 = df[(df['Datetime'] >= a) & (df['Datetime'] <= b)]
```

```python
df1 = df1.drop(columns=['Date','Time'])
```

```python
df1
```

### 3. Create a Histogram

```python
plt.figure(figsize=(8,6))
sns.histplot(df1['Voltage'], bins = 100,kde=True, color='red')
plt.title('Distribution of Voltage')
plt.xlabel('Voltage (V)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

sns.histplot(df1['Global_reactive_power'], bins=20, kde=True, color='blue', ax=axes[0])
axes[0].set_title('Global Reactive Power Distribution')
axes[0].set_xlabel('Global Reactive Power (kW)')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

sns.histplot(df1['Global_active_power'], bins=20, kde=True, color='green', ax=axes[1])
axes[1].set_title('Global Active Power Distribution')
axes[1].set_xlabel('Global Active Power (kW)')
axes[1].set_ylabel('Frequency')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### 4) Timeseries Plot

```python
df1.set_index('Datetime', inplace=True)
df1
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(24, 54), sharex=True)

sns.lineplot(x=df1.index, y=df1['Global_active_power'], color='green', ax=axes[1])
axes[1].set_title('Global Active Power Over Time')
axes[1].set_xlabel('Datetime')
axes[1].set_ylabel('Global Active Power (kW)')
axes[1].grid(True)

sns.lineplot(x=df1.index, y=df1['Global_reactive_power'], color='blue', ax=axes[0])
axes[0].set_title('Global Reactive Power Over Time')
axes[0].set_xlabel('Datetime')
axes[0].set_ylabel('Global Reactive Power (kW)')
axes[0].grid(True)

plt.tight_layout()
plt.show()
```

### 5) Plots for submetering

```python
if df1.index.duplicated().any():
    df1 = df1[~df1.index.duplicated(keep='first')]
    
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(24, 54), sharex=True)

sns.lineplot(x=df1.index,y=df1['Sub_metering_1'],color='red',ax=axes[0])
axes[0].set_title('Submetering 1, Over Time')
axes[0].set_ylabel('Submetering 1')
axes[0].set_xlabel('Datetime')
axes[0].grid(True)

sns.lineplot(x=df1.index,y=df1['Sub_metering_2'],color='blue',ax=axes[1])
axes[1].set_title('Submetering 2, Over Time')
axes[1].set_ylabel('Submetering 2')
axes[1].set_xlabel('Datetime')
axes[1].grid(True)

sns.lineplot(x=df1.index, y=df1['Sub_metering_3'], color='green', ax=axes[2])
axes[2].set_title('Sub Metering 3 Over Time')
axes[2].set_xlabel('Datetime')
axes[2].set_ylabel('Sub Metering 3')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

### 6) Create Different Kinds of Plots !!

```python
df1.columns
```

#### Countplot

```python
plt.figure(figsize=(24, 6))
sns.countplot(x='Global_intensity', data=df1, palette='coolwarm')
plt.title('Count Plot of Global Intensity Levels')
plt.xlabel('Global Intensity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df1.dtypes
```

#### Heatmap

```python
for column in ['Global_active_power', 'Global_reactive_power', 'Voltage','Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']:
    df1[column] = pd.to_numeric(df1[column], errors='coerce')

df1.fillna(df1.mean(), inplace=True)


correlation_matrix = df1[['Global_active_power', 'Global_reactive_power', 'Voltage','Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].corr()

plt.figure(figsize=(16, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

#### Jointplot

```python
sns.jointplot(x='Sub_metering_3', y='Global_active_power', data=df1, kind='scatter', color='purple')
plt.suptitle('Joint Plot of Sub_metering 3 and Global active power', y=1.02)
plt.show()
```

```python
sns.jointplot(x='Sub_metering_3', y='Global_intensity', data=df1, kind='reg', color='red')
plt.suptitle('Joint Plot of Sub_metering 3 and Global_intensity', y=1.02)
plt.show()
```

#### Distplot

```python
plt.figure(figsize=(8, 6))
sns.histplot(df1['Voltage'], bins=30, kde=True, color='brown')
plt.title('Distribution Plot of Voltage')
plt.xlabel('Voltage (V)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df1['Global_intensity'], bins=30, kde=True, color='Blue')
plt.title('Distribution Plot of Global Intensity')
plt.xlabel('Global Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Piechart

```python
plt.figure(figsize=(10, 6))
sub_metering_totals = df1[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].sum()
labels=['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
colors = sns.color_palette('bright')[6:9]

plt.pie(sub_metering_totals, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

#### Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1[['Global_active_power', 'Global_reactive_power']], palette='bright')
plt.title('Box Plot of Global Active and Reactive Power')
plt.ylabel('Power(kw)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Voltage', y='Global_intensity', data=df1, color='red',markers="*")
plt.title('Scatter Plot of Voltage vs Global Intensity')
plt.xlabel('Voltage (V)')
plt.ylabel('Global Intensity (A)')
plt.grid(True)
plt.show()
```
