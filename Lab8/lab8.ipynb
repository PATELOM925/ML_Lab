```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
print(df['Species'].unique())
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['encoding'] = label_encoder.fit_transform(df['Species'])
df.head()
```

```python
df_X = df.iloc[:, 1:5]
df_y = df.iloc[:, -1:]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_X_scaled = scaler.fit_transform(df_X)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR

estimator = SVR(kernel="linear")
selector1 = RFE(estimator, n_features_to_select=5, step=1)
selector1 = selector1.fit(df_X_scaled, df_y)
print(selector1.ranking_)
print(selector1.support_)
```

```python
# Feature Selection Method1 

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

selector = SelectFromModel(RandomForestClassifier(n_estimators=100))
X_new = selector.fit_transform(df_X_scaled, df_y)
X_new.shape
```
